```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#bring data in from sample census survey data
#look at first few rows to see what we're looking at
df_census = pd.read_csv('ACS_16_1YR_R0201.US01PRF.csv')
df_census.head()
```

```python
%matplotlib inline
df_census2 = df_census[['GRT_STUB.display-label', 'EST']].set_index('GRT_STUB.display-label')
```

```python
df_census2.plot(kind='bar', figsize=(20,10))
```

The data above shows the percentage of people who identify as white in a given state relative the state's overall population. I don't think this data is particularly interesting, but it was one of the first pieces of information I found in the American survey data that had multiple rows rather than just showing an annual aggregation.

```python
# bring in World Cup information which has a lot more interesting options
df_wcup = pd.read_csv('WorldCups.csv')
df_wcup.head()
```

```python
#create a new df with just the country and their number of occurrences
df_winner = df_wcup['Country'].value_counts()

df_winner.plot(kind='bar', figsize = (10,4.5), color='purple')
plt.ylabel('# of Championships')
plt.show()
```

A simple graph to show how many times each country in the sample list has won the World Cup. 

```python
# reduce some of the columns to create a plot chart
df_wcup2 = df_wcup[['Year', 'GoalsScored']].set_index('Year')
#df_wcup2.plot(kind='bar', figsize=(5,5), color='green')

plt.subplot(1,2,1)
plt.plot(df_wcup2)
plt.ylabel('Goals Scored')

df_wcup3 = df_wcup[['Year', 'MatchesPlayed']].set_index('Year')
#df_wcup3.plot(kind='bar', figsize=(5,5), color='green')
plt.subplot(1,2,2)
plt.plot(df_wcup3)
plt.ylabel('Matches Played')

plt.show()
```

This graph shows the total number of goals scored for each World Cup series in the years identified. Additional analysis could be interesting to determine why goals are trending upward. The second graph shows the total number of matches played by world cup year, which could account for some of the increase in goals scored. However, there is still a fair amount of variance in goals scored within like number of matches.

```python
# create a new column with data: attendance per match, goals scored per match played
df_wcup['Goals per Match'] = df_wcup['GoalsScored'] / df_wcup['MatchesPlayed']
df_wcup4 = df_wcup[['Year', 'Goals per Match']].set_index('Year')

# plot the new column with a bar graph
df_wcup4.plot(kind='bar', figsize = (10,5), color='red')
plt.ylabel('Goals per Match')
plt.show()
```

Despite the number of goals in aggregate trending upward, this actually shows that the number of goals per match is actually slightly steady to declining every year. It would be interesting to look further into the data set and determine why this might be the case. Another additional insight would be to look at some stats of this data set to see the average once maybe some outliers like 1954 were removed as they look to skew the data.
