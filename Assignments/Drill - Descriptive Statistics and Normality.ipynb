```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
gamma = np.random.gamma(5,1, 100)

# function to graph histogram and plot lines
def graphing (gamma):
    plt.hist(gamma)
    plt.axvline(gamma.mean(), color='b', linestyle='solid', linewidth=2)
    plt.axvline(gamma.mean() + gamma.std(), color='b', linestyle='dashed', linewidth=2)
    plt.axvline(gamma.mean() - gamma.std(), color='b', linestyle='dashed', linewidth=2) 
    plt.show()
graphing(gamma)

def stats (gamma):
    print("mean: ",np.mean(gamma))
    print("st.dev.: ",np.std(gamma))
stats(gamma)
```

The mean is not helpful in this case, but the standard deviation might be. The majority of the occurences are less than the mean with some extreme big number outliers skewing the data. The standard deviation would be more helpful as it looks like the majority of the data is within 1 standard deviation. However, there is significantly more data in the second quartile and is not distributed normally around the mean.

```python
beta = np.random.beta(10,20,100)

graphing(beta)
stats(beta)
```

This looks more normal and therefore the mean and standard deviation would be more helpful in understanding the distribution. There are still some low outliers skewing the information, and there is much more data in the 2nd quartile as well. 

```python
triangular = np.random.triangular(10,15,20,100)

graphing(triangular)
stats(triangular)
```

This looks very similar to the graph above. I think it would be fairly useful to look at the mean and standard deviation.

```python
exponential = np.random.exponential(40, 100)

graphing(exponential)
stats(exponential)
```

Graphing an exponential function like this is very different from a normal distribution, the mean and standard deviation aren't helpful at all. 

```python
lognormal = np.random.lognormal(40,20,100)

graphing(lognormal)
stats(lognormal)
```

This function produced most, if not all, of its values close to the mean, but it still isn't normally distributed. You could argue that the mean and standard deviation do explain a lot of the graph as all the values are within 1 standard deviation, but I'm not sure that's truly explaining what's going on with the lognormal.

```python
logistic = np.random.logistic(5,1,100)

graphing(logistic)
stats(logistic)
```

This graph seems to fairly normal. The mean and standard deviation would be useful in explaining the distribution. However, there are some outlier values in the lower quartiles that are slightly skewing the data, which seems to be offset by a large volume of values in the third quartile.

```python
norm1 = np.random.normal(5,0.5,100)
norm2 = np.random.normal(10,1,100)

norm3 = norm1 + norm2
graphing (norm3)
stats(norm3)
```

Because there is a small sample size of 100 occurences the data is slightly skewed. I'm surprised at how skewed the data is actually, and makes me wonder if some of the graphs above would be more normal with a higher sample size. There's also probably something to do with adding two normal distributions to each other.
